```python
import pmdarima as pm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sp500sub.csv', index_col='Date', parse_dates=True)
```

```python
df.head()
```

```python
goog = df[df['Name'] == 'GOOG']['Close']
# goog = np.log(df[df['Name'] == 'GOOG']['Close'])
```

```python
goog.plot();
```

```python
Ntest = 30
train = goog.iloc[:-Ntest]
test = goog.iloc[-Ntest:]
```

```python
model = pm.auto_arima(train, trace=True, error_action='ignore', seasonal=False, max_p = 10, max_q=3, max_order=13, suppress_warnings=True, stepwise=False)
```

```python
model.summary()
```

```python
model.get_params()
```

```python
def plot_result(model, fulldata, train, test):
  p = model.get_params()
  d = p['order'][1]
  train_pred = model.predict_in_sample(start=d, end=-1)
  test_pred, confint = model.predict(n_periods=Ntest, return_conf_int=True)
  fig, ax = plt.subplots(figsize=(10, 5))
  ax.plot(fulldata.index, fulldata, label='data')
  ax.plot(train.index[d:], train_pred, label='fitted')
  ax.plot(test.index, test_pred, label='forecast')
  ax.fill_between(test.index, \
                  confint[:,0], confint[:,1], \
                  color='red', alpha=0.3)
  ax.legend();
```

```python
plot_result(model, goog, train, test)
```

```python
def plot_test(model, test):
  test_pred, confint = model.predict(n_periods=Ntest, return_conf_int=True)
  fig, ax = plt.subplots(figsize=(10, 5))
  ax.plot(test.index, test, label='true')
  ax.plot(test.index, test_pred, label='forecast')
  ax.fill_between(test.index, \
                  confint[:,0], confint[:,1], \
                  color='red', alpha=0.3)
  ax.legend();
```

```python
plot_test(model, test)
```

```python
def rmse(y, t):
  return np.sqrt(np.mean((t-y)**2))
```

```python
#print(model.predict(Ntest))
#print(test)
print("RMSE ARIMA:", rmse(model.predict(n_periods=Ntest).values, test.values))
print("RMSE Naive:", rmse(train.iloc[-1], test))
```

```python
#aapl = df[df['Name'] == 'AAPL']['Close']
aapl = np.log(df[df['Name'] == 'AAPL']['Close'])
aapl.plot();
```

```python
train = aapl.iloc[:-Ntest]
test = aapl.iloc[-Ntest:]
```

```python
model = pm.auto_arima(train, maxiter=20, suppress_warnings=True, seasonal= False, trace=True, error_action='ignore')
```

```python
model.summary()
```

```python
plot_result(model, aapl, train, test)
```

```python
plot_test(model, test)
```

```python
print("RMSE ARIMA:", rmse(np.exp(model.predict(Ntest).values), np.exp(test.values)))
print("RMSE Naive:", rmse(np.exp(train.iloc[-1]), np.exp(test)))
```
