```python
x = np.linspace(0, 10, 1000)  # X values from 0 to 10
y = 1.2 * x + np.random.normal(scale=1, size=len(x))  # Linear trend + noise
coeffs = np.polyfit(x, y, deg=1)
trendline = np.poly1d(coeffs)
plt.scatter(x, y, s=10)
plt.plot(x, trendline(x), color="red", linewidth=2, label="Trendline")
plt.show()
```

```python
from scipy.stats import multivariate_normal
```

```python
mean = np.array([0, 0])
cov = np.array([[1, -0.5], [-0.5, 2]])
mvn = multivariate_normal(mean=mean, cov=cov)
samples_mvn = mvn.rvs(size=1000)
#print(samples_mvn)
```

```python
plt.scatter(samples_mvn[:, 0], samples_mvn[:, 1], alpha=1, s=10)
plt.show()
```

```python
mean_estimated = np.mean(samples_mvn, axis=0)  # Mean of each column
cov_estimated = np.cov(samples_mvn.T)  # Covariance matrix,.T is the transpose
print(f"Mean: {mean_estimated}, Covariance: {cov_estimated}")
```

```python
# The formulas for the estimations are the following:
size = 1000
mu = (1/size) * np.sum(samples_mvn, axis=0)
center =  samples_mvn - mu
cov = (center.T @ center) / (size - 1)
print(f"Mean: {mu}, Covariance: {cov}")
```
